```python
def fib(n): 
    if n == 0 or n == 1:
        return 1
    n0 = 1
    n1 = 1
    for _ in range(n+1):
        n0, n1 = n1, n0+n1
    return n1
```

```python
mu = 100  
sigma = 15
x = np.random.normal(mu, sigma, (5000,))
# x = np.random.uniform(0, 1, (10000,))
```

```python
plt.scatter(np.linspace(0,1, x.shape[0]), x, marker='.')
plt.plot([0, 1], [100, 100], 'r')
```

```python


# num_bins = 50
num_bins = 100
# num_bins = 3
# num_bins = 20

fig, ax = plt.subplots()
n, bins, _ = ax.hist(x, num_bins, density=True)
n

scale = (1 / (np.sqrt(2 * np.pi) * sigma))
x1 = 1 / sigma * (bins - mu)
y =  scale * np.exp(-0.5 * (x1)**2)

ax.plot(bins, y, '--')
ax.set_xlabel('X')
ax.set_ylabel('Probability density')
ax.set_title(r'Histogram: $\mu=100$, $\sigma=15$')

# # Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
# plt.show()
```
